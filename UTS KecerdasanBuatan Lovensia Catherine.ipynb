**Daftar Semua Karyawan Berdasarkan Domain Perusahaannya**

**Fungsi yang Mengembalikan Perusahaan Beserta Jumlah Karyawan Dari Berbagai Negara yang Bekerja di Perusahaan Tersebut**

```python
from collections import defaultdict

def get_company_employee_nationality_count():
    company_nationality_counts = defaultdict(lambda: defaultdict(int))

    for employee in employee_detail_list:
        company_name = employee['company']
        nationality = employee['nationality']
        company_nationality_counts[company_name][nationality] += 1

    return [
        {
            "company": company,
            "employee_nationality": dict(nationality_counts)
        }
        for company, nationality_counts in company_nationality_counts.items()
    ]

result = get_company_employee_nationality_count()
print(result)
```
